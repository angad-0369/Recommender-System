```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Coursera.csv')
df.head()
```

```python
df['Course Rating'].unique()
```

```python

df['Course Rating']=df['Course Rating'].replace(':',' ')
df['Course Rating']=df['Course Rating'].replace(',',' ')

df['Course Description']=df['Course Description'].replace('-',' ')
df['Course Description']=df['Course Description'].replace('.',' ')
df['Course Description']=df['Course Description'].replace(':',' ')
df['Course Description']=df['Course Description'].replace(',',' ')

df['Skills']=df['Skills'].replace('-',' ')
df['Skills']=df['Skills'].replace('.',' ')
df['Skills']=df['Skills'].replace(':',' ')
df['Skills']=df['Skills'].replace(',',' ')
df['Skills']=df['Skills'].replace('(',' ')
df['Skills']=df['Skills'].replace(')',' ')
```

```python
df = df[df['Course Rating'] != 'Not Calibrated']
df['Course Rating'] = df['Course Rating'].astype(float)
df
```

```python
df.drop('Course URL',axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf = TfidfVectorizer(stop_words='english')

df['Course Description'] = df['Course Description'].fillna('')


tfidf_matrix = tfidf.fit_transform(df['Course Description'])

```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_matrix)
```

```python
similarity
```

```python
indices = pd.Series(df.index, index=df['Course Name']).drop_duplicates()
```

```python
def get_recommendations(title, cosine_sim=similarity):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    courses = [i[0] for i in sim_scores]

    return df['Course Name'].iloc[courses]
```

```python
get_recommendations('Finance for Managers')
```

```python
get_recommendations('Silicon Thin Film Solar Cells')
```
